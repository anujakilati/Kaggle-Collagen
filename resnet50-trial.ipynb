```python
def sample_minority(df, labelType):
  sample = df[df['label'] == labelType]
  X_train = sample.drop('fitzpatrick_scale', axis=1)
  y_train = sample['fitzpatrick_scale']
  ros = RandomOverSampler(sampling_strategy="auto", random_state=42)
  X_result, y_result = ros.fit_resample(X_train, y_train)
  resampled_data = pd.concat([pd.DataFrame(X_result, columns=X_train.columns), pd.Series(y_result, name='fitzpatrick_scale')], axis=1)
  return resampled_data
```

```python
resampled_fitzpatrickScale = pd.DataFrame()
for i in sample_fitzpatrickScale['label'].unique():
  resampled_fitzpatrickScale = pd.concat([resampled_fitzpatrickScale, sample_minority(sample_fitzpatrickScale, i)], ignore_index=True)
```

# Test with Image Data Generator

**Histogram Equalization**: https://sigmoidal.ai/en/histogram-equalization-with-opencv-and-python/

```python
def histogram_equalization(img):
    hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)  
    h, s, v = cv2.split(hsv_img)  
    v_eq = cv2.equalizeHist(v)  
    hsv_eq = cv2.merge([h, s, v_eq]) 
    img_eq = cv2.cvtColor(hsv_eq, cv2.COLOR_HSV2RGB) 
    return img_eq
```

**Clahe**

```python
def apply_clahe(img):
    image_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    image_hsv[:, :, 2] = clahe.apply(image_hsv[:, :, 2])
    image_equalized = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2BGR)
    return image_equalized
```

# Transfer Learning: Xception Model

**add class weight** 

```python
from sklearn.utils.class_weight import compute_class_weight
labels = train_generator.classes  

class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)

class_weight_dict = {i: w for i, w in enumerate(class_weights)}
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization, Rescaling

# Load Pretrained ResNet-50 (closest alternative to ResNet-18)
base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze pre-trained layers

# Define Input Layer
inputs = keras.Input(shape=(224, 224, 3))

# Normalize Input
x = Rescaling(1.0 / 255)(inputs)  

# Feature Extraction
x = base_model(x, training=False)  
x = GlobalAveragePooling2D()(x)  

# Add Fully Connected Layers
x = BatchNormalization()(x)
x = Dense(512, activation='relu', kernel_regularizer=keras.regularizers.l2(0.001))(x)
x = Dropout(0.2)(x)
x = BatchNormalization()(x)
x = Dense(256, activation='relu', kernel_regularizer=keras.regularizers.l2(0.001))(x)
x = Dropout(0.2)(x)

# Output Layer (21 classes)
outputs = Dense(21, activation="softmax")(x)

# Create Model
model = keras.Model(inputs, outputs)

# Show Model Summary
model.summary()
```

```python
base_model.trainable = True
model.summary(show_trainable=True)

num_classes = 21
f1_metric = tf.metrics.F1Score(average='weighted')

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),
    loss=keras.losses.CategoricalCrossentropy(from_logits=False),
    metrics=["accuracy", f1_metric],
)

epochs = 40

print("Fitting the end-to-end model")
#history = model.fit(train_generator, epochs=epochs, validation_data=val_generator)
history = model.fit(train_generator, epochs=epochs, validation_data=val_generator)
```

# Plot Training and Validation losss and accuracies

```python
plt.plot(range(1, epochs + 1), history.history['loss'], label='Training Loss')
plt.plot(range(1, epochs + 1), history.history['val_loss'], label='Validation Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()


# Plot training and validation accuracy
plt.plot(range(1, epochs + 1), history.history['accuracy'], label='Training Accuracy')
plt.plot(range(1, epochs + 1), history.history['val_accuracy'], label='Validation Accuracy')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
